## Lower Data

```python
data=data.astype(str)
data["Label"] = data["Label"].apply(lambda x: x.lower())
data["Judul"] = data["Judul"].apply(lambda x: x.lower())
data["abAtrak"] = data["Abstrak"].apply(lambda x: x.lower())
```

## Steaming

```python
!pip install Sastrawi
```

```python
# import Sastrawi package
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stem
data1 = data
for i in range(len(data1)):
  abstrak = data1["Abstrak"][i]
  output   = stemmer.stem(abstrak)
  data1["Abstrak"][i] = output
```

```python
print(len(data1))
```

## Tokenizing

```python
data1=data1.astype(str)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Inisialisasi fungsi stop words
stop_factory = StopWordRemoverFactory()

words = []

# Membuat perulangan untuk memasukkan dataset ke dalam tekonisasi dan list stopwords
for i in range (len(data1)):
  # Inisialisai fungsi tokenisasi dan stopword
  tokens = word_tokenize(data1["Abstrak"][i])
  more_stopword = ['dengan', 'ia','bahwa','oleh','aalysis','aam','kunci']
  data = stop_factory.get_stop_words()+more_stopword
  stopword = stop_factory.create_stop_word_remover()

  # removed kata
  removed = []
  for t in tokens:
      if t not in data:
          removed.append(t)
  
  # Memasukkan hasil removed kedalem variable words
  words.append(removed)
  print(removed)
```

## Penggabungan kata

```python
gabung=[]
for i in range(len(words)):
  joinkata = ' '.join(words[i])
  gabung.append(joinkata)

result = pd.DataFrame(gabung, columns=['abstrak'])
result
```

## Melakukan ekstraksi fitur dengan TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

### Count Vacktorizer

```python
countvectorizer = CountVectorizer(analyzer= 'word', stop_words='english')
count_wm = countvectorizer.fit_transform(gabung)
count_tokens = countvectorizer.get_feature_names_out()
df_countvect = pd.DataFrame(data = count_wm.toarray(),columns = count_tokens)
print("Count Vectorizer\n")
df_countvect
```

### Vecktorizer

```python
# TfidfVectorizer 
tfidfvectorizer = TfidfVectorizer(analyzer='word',stop_words= 'english')
tfidf_wm = tfidfvectorizer.fit_transform(gabung)
tfidf_tokens = tfidfvectorizer.get_feature_names_out()
df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),columns = tfidf_tokens)
print("\nTF-IDF Vectorizer\n")
df_tfidfvect
```

## Reduksi dimensi menggunakan PCA

```python
from sklearn.decomposition import PCA
pca_abstrak = PCA(n_components=30) #Nilai n = 30
principalComponents_abstrak = pca_abstrak.fit_transform(df_tfidfvect) #Transform ke dalam data tf-idf
principal_abstrak_Df = pd.DataFrame(principalComponents_abstrak)
principal_abstrak_Df
```

## Split data

```python
from sklearn.model_selection import train_test_split
xtrain, xtest = train_test_split(principalComponents_abstrak,test_size=0.2, random_state=1)#Nilai X training dan Nilai X testing
ytrain, ytest = train_test_split(data1["Label"], test_size=0.2, random_state=1)#Nilai Y training dan Nilai Y testing
```

## Modelling

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(xtrain, ytrain)
```

```python
test_pred = modelKNN.predict(xtest)
test_pred
```

```python
accuracy_score(ytest, test_pred)
```

### Naive Bayes

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
gaussian.fit(xtrain, ytrain)
```
