```python
!pip install Twint
!pip install snscrape
import twint
import pandas as pd
import snscrape.modules.twitter as sracp
```

```python
topik = "ganjar pranowo"
tweets = []
limit = 100
for tweet in sracp.TwitterSearchScraper(topik).get_items():
  if limit==0:
    break
  else:
    tweets.append([tweet.content,tweet.user,tweet.place])
    limit-=1
df = pd.DataFrame(tweets, columns=["content","username","lokasi"])
df
```

```python
df.to_csv("twitter.csv",index=False)
```

## Web PTA Trunojoyo

```python
!pip install requests
!pip install beautifulsoup4
# !pip install tweepy
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def pta():
  data={"nama":[],"judul":[],"abstrak":[]}
  for i in range(1,4):
    url = "https://pta.trunojoyo.ac.id/c_search/byprod/10/{}".format(i)
    r = requests.get(url)
    request = r.content
    soup = BeautifulSoup(request,"html.parser")
    artikels = soup.findAll('li', attrs={'data-cat':'#luxury'})
    for artikel in artikels:
      respone = requests.get(artikel.find('a', class_='gray button')['href'])
      soup1 = BeautifulSoup(respone.content,"html.parser")
      isi = soup1.find('div', attrs={'id':'content_journal'})
      nama = isi.findAll('div')[1].span.text.split(' : ')[1]
      judul = isi.findAll('a',attrs={'class':'title'})[0]
      abstrak = isi.findAll('p',attrs={'align':'justify'})[0]
      data["nama"].append(nama)
      data["judul"].append(judul.text)
      data["abstrak"].append(abstrak.text)
  df=pd.DataFrame(data)
  df.to_csv("pta.csv",index=False)
  return df
# df
```

## Web

```python
def web():
  data={"judul":[],"isi berita":[]}
  for i in range(1,10):
    url = "https://www.pikiran-rakyat.com/indeks-berita?page={}".format(i)
    r = requests.get(url)
    request = r.content
    soup = BeautifulSoup(request,"html.parser")
    artikels = soup.findAll('div', attrs={'class':'latest__item'})
    for artikel in artikels:
      # respone = requests.get(artikel.find('a', class_='gray button')['href'])
      respone = requests.get(artikel.find("a",attrs={'class':'latest__link'})['href']+"?page=all")
      # print(artikel.find("a",attrs={'class':'latest__link'})['href'])
      soup1 = BeautifulSoup(respone.content,"html.parser")
      isi = soup1.find('article', attrs={'class':'read__content clearfix'})
      judul = soup1.find('h1',attrs={'class':'read__title'})
      # abstrak = isi.find('p')
      # print(judul)
      abstrak=' '.join(isi.stripped_strings)
      # print(' '.join(isi.stripped_strings))
      data["judul"].append(judul.text)
      data["isi berita"].append(abstrak)
  # print(data)
  df=pd.DataFrame(data)
  df.to_csv("web.csv",index=False)
  return df
```

```python
web()
```
